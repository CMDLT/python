### 完美原始折线图1（辅助显示器）

#### 案例：显示温度变化状况
- 需求：画出某城市11点到12点1小时内每分钟的温度变化折线图，温度范围在15度~18度

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#准备数据
x=range(60)
y_shanghai=[np.random.uniform(15,18) for i in x]#15到18均匀的产生数据
#开始画图三步走
plt.figure(figsize=(17,6),dpi=(100))
plt.plot(x,y_shanghai)

#构造x、y轴刻度标签
x_ticks_label=['11点{}分'.format(i) for i in x]
#修改x的刻度
plt.xticks(x[::5],x_ticks_label[::5])
plt.yticks(range(0,40,5))

#添加网格显示
plt.grid(linestyle='--',alpha=0.5)

#添加描述信息
plt.xlabel('时间变化')
plt.ylabel('温度变化')
plt.title('上海11点到12点每分钟的温度变化状况')

plt.show()
```

### 2.2.3 完美原始折线图2（图像层）

### 2.3 散点图

### 2.3.1 常见图形的种类及意义

###### 折线图plot：以折线的上升或者下降来表示统计数量的增减或变化的统计图
    - 特点：能够显示数据的变化趋势，反应事物的变化情况
    
###### 散点图scatter：用两组数据构成多个坐标点，考察坐标点的分布，判断两变量之间是否存在某种关联或总结坐标点的分布模式
    - 特点：判断变量之间是否存在关联趋势，展示离群点（分布规律）
    
###### 柱状图bar：排列在工作表中的列或者行的数据可以绘制到柱状图中
    - 特点：绘制离散的数据，能够一眼看出各个数据的大小，比较数据之间的差别。（统计/对比）
    
###### 直方图histogram：由一系列高度不等的纵向条纹或线段表示数据分布的情况。一般用横轴表示数据范围，纵轴表示分布情况
    - 特点：绘制连续性的数据展示一组或多组数据的分布情况（统计）
    
###### 饼图pie：用于表示不同分类的占比情况，通过弧度大小来对比各种分类
    - 特点：分类数据的占比情况

### 2.4.1 柱状图绘制

- 需求：对比每部电影票房收入

```python
import matplotlib.pyplot as plt
# 1、准备数据
movie_name = ['雷神3','正义联盟','东方快车谋杀案','寻梦环游记','全球风暴','降魔传','追捕','ike','vox']
tickets = [73853,57767,22354,15969,14839,8725,8716,8318,6916]

# 2、创建画布
plt.figure(figsize=(10,5),dpi=100)

# 3、绘制柱状图
# matplotlib.pyplot.bar(x,width,align='center',**kwargs)
x_ticks = range(len(movie_name))
plt.bar(x_ticks,tickets,color=('#0f3b8b','#173f87','#234688','#2d4d88','#365997','#4265a5','#5072b0','#5575b0','#6a8ac6'))

# 修改x刻度
plt.xticks(x_ticks,movie_name)

# 添加标题
plt.title('电影票房收入')

# 添加网格显示
plt.grid(linestyle='--',alpha=0.3)

# 4、显示图像
plt.show()
```

### 2.5.1 直方图
- 相关概念
    - 组数：在统计数据时，我们把数据按照不同的范围分成几个组，分成的组的个数称为组数
    - 组距：每一组两个端点的差
- 直方图与柱状图的差别
    - 1、直方图展示数据的分布，柱状图比较数据的大小
    - 2、直方图x轴为定量数据，柱状图x轴为分类数据
    - 3、直方图柱子无间隔，柱状图柱子有间隔
    - 4、直方图柱子宽度可不一，柱状图柱子宽度须一致

### 2.5.3 直方图绘制
- y轴的变量可以是频次（数据出现了多少次）、频数（频次/总次数），频率/组距，不同的变量会让直方图描述的数据分布意义不同

```python
import matplotlib.pyplot as plt
# 需求：电影时长分布状况
# 1、准备数据
time = [121,88,99,90,97,111,100,109,128,23,123,87,90,124,127,95,80,70,40,160,111,100,120,130,140,150,
        45,78,75,60,66,88,99,67,180,222,109,135,145,145,146,134,124,145,132,156,87,89,96,78,89,123,139]

# 2、创建画布
plt.figure(figsize=(10,5),dpi=100)

# 3、绘制直方图
# matplotlib.pyplot.hist(x,bins=None,density=None,**kwargs) # bins是组数，bins 组数 = (max(time) - min(time)) // 组距
# density是是否显示频率（默认为否）
distance = 8 # 另组距为2
group_num = int((max(time) - min(time)) / distance) #除用‘//’外，也可用int()强转为整形
plt.hist(time,bins=group_num,density=True)

# 修改x轴刻度
plt.xticks(range(min(time),max(time)+2,distance)) # 从最小值显示到最大值,因为x轴数据少了一个格子，所以我们加一个2，步长设为组距

# 添加网格
plt.grid(linestyle='--',alpha=0.3)

# 添加标题
plt.title('电影时长分布情况')

# 4、显示图像
plt.show()
```

### 2.3.4 直方图的应用场景
- 用于表示分布情况
- 通过直方图还可以观察和估计哪些数据比较集中，异常或孤立的数据分布在何处
- 例如：用户年龄分布，商品价格分布

### 2.6.2 饼图绘制

```python
# 1、准备数据
movie_name = ['雷神3','正义联盟','东方快车谋杀案','寻梦环游记','全球风暴','降魔传','追捕','ike','vox']
place_count = [73853,57767,22354,15969,14839,8725,8716,8318,7916]

# 2、创建画布
plt.figure(figsize=(10,5),dpi=100)

# 3、绘制饼图
# plt.pie(x,labels= ,autopct= ,colors)
# x：数量，自动算百分比
# labels：每部分名称
# autopct：占比显示指定%1.2f%%（%1.2f是占1位，保留2位，%%显示%，多加一个%表转义）
# colors：每部分颜色
plt.pie(place_count,labels=movie_name,colors=('#0f3b8b','#173f87','#234688','#2d4d88','#365997','#4265a5','#5072b0','#5575b0','#6a8ac6'),
        autopct="%1.2f%%")

# 显示图例
plt.legend()

# 调整图形长高相同，且位置居中
plt.axis('equal')

# 4、显示图像
plt.show()
```

### 饼图应用场景
- 分类的占比情况（不超过9个分类）
- 例如：班级男女分布占比，公司销售额占比
